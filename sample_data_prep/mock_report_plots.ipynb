```python
import matplotlib.pyplot as plt
import numpy as np

# Simulated influencer data
influencers = ['İrem', 'İlkay', 'Özgür', 'Burak', 'Mehmet']
sales = np.random.randint(50, 300, size=len(influencers))

plt.figure(figsize=(10, 6))
bars = plt.bar(influencers, sales, color='mediumseagreen', edgecolor='black')

# Add value labels
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, yval, ha='center', fontsize=10)

plt.title('En Çok Ürün Sattıran Influencerlar', fontsize=14)
plt.ylabel('Satış Sayısı')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
plt.savefig('influencer_sales.png', dpi=300, bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt

roles = ['Analyst', 'Product Manager', 'Campaign Manager', 'Influencer Manager', 'Finance Manager', 'Admin']
counts = np.random.randint(2, 10, size=len(roles))

plt.figure(figsize=(8, 8))
plt.pie(counts, labels=roles, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
plt.title('User Rol Dağılımı', fontsize=14)
plt.show()
plt.savefig('user_role_distribution.png', dpi=300, bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt

years = np.arange(2020, 2025)
influencer_a = np.random.randint(10, 100, size=len(years))
influencer_b = np.random.randint(20, 120, size=len(years))

plt.figure(figsize=(10, 5))
plt.plot(years, influencer_a, marker='o', label='Influencer İrem', color='dodgerblue')
plt.plot(years, influencer_b, marker='s', label='Influencer İlkay', color='orange')

plt.title('Yıl Bazında Influencerların Getirdiği Müşteri Sayısı')
plt.xlabel('Yıl')
plt.ylabel('Kullanıcı Sayısı')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
plt.savefig('influencer_customers.png', dpi=300, bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt

campaigns = ['Bahar İndirimi', 'Yaz Çılgınlığı', 'Harika Cuma', 'Yeni Yıl', 'Okula Dönüş']
budgets = np.random.uniform(5000, 20000, size=len(campaigns))

plt.figure(figsize=(10, 6))
bars = plt.barh(campaigns, budgets, color='salmon', edgecolor='black')
plt.xlabel('Bütçe (₺)')
plt.title('En Yüksek Bütçeli Kampanyalar')

# Add value labels
for i, bar in enumerate(bars):
    plt.text(bar.get_width() + 200, i, f"{bar.get_width():,.0f}", va='center')

plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
plt.savefig('campaign_budgets.png', dpi=300, bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt

influencers = ['İrem', 'İlkay', 'Özgür', 'Burak']
likes = np.random.randint(50, 200, size=len(influencers))
sales = np.random.randint(30, 180, size=len(influencers))

x = np.arange(len(influencers))
width = 0.35

plt.figure(figsize=(10, 6))
plt.bar(x - width/2, likes, width, label='Beğeni', color='skyblue')
plt.bar(x + width/2, sales, width, label='Satış', color='lightgreen')

plt.xlabel('Influencer')
plt.ylabel('Adet')
plt.title('Influencer Bazlı Beğeni ve Satış Karşılaştırması')
plt.xticks(x, influencers)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
plt.savefig('influencer_sales_likes.png', dpi=300, bbox_inches='tight')
```

```python
categories = ['Elektronik', 'Kozmetik', 'Moda', 'Ev & Yaşam']
likes = np.random.randint(100, 300, size=len(categories))
sales = np.random.randint(150, 350, size=len(categories))

x = np.arange(len(categories))

plt.figure(figsize=(10, 6))
plt.bar(x, likes, label='Beğeni', color='lightblue')
plt.bar(x, sales, bottom=likes, label='Satış', color='orange')

plt.xticks(x, categories)
plt.ylabel('Toplam')
plt.title('Kategori Bazlı Beğeni ve Satış')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
plt.savefig('category_sales_likes.png', dpi=300, bbox_inches='tight')
```

```python
import numpy as np
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales_ilkay = np.random.randint(10, 50, len(months))
sales_ozgur = np.random.randint(20, 60, len(months))
sales_burak = np.random.randint(5, 45, len(months))
sales_irem = np.random.randint(15, 55, len(months))

plt.figure(figsize=(10, 6))
plt.plot(months, sales_ilkay, label='İlkay', marker='o')
plt.plot(months, sales_ozgur, label='Özgür', marker='s')
plt.plot(months, sales_burak, label='Burak', marker='^')
plt.plot(months, sales_irem, label='İrem', marker='d')

plt.title('Kampanya Bazlı Aylık Satış Trendleri')
plt.xlabel('Ay')
plt.ylabel('Satış Miktarı')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
plt.savefig('sales_trends.png', dpi=300, bbox_inches='tight')
```
